Geef daarbij per medewerker in een aparte kolom zowel ‘kort in dienst’ als ‘lang in dienst’ aan. Gebruik (wederom) een for-loop.<br>
(2 kolommen, 102 rijen) 

```python
from datetime import date

current_year = date.today().year

kort_in_dienst = []
lang_in_dienst = []

for start_date in sales_staff['DATE_HIRED']:

    years_in_service = current_year - current_year
    
    # Check if the employee has been in service for less than 25 years
    if years_in_service < 25:
        kort_in_dienst.append('Yes')
    else:
        kort_in_dienst.append('No')
    
    # Check if the employee has been in service for 12 years or more
    if years_in_service >= 12:
        lang_in_dienst.append('Yes')
    else:
        lang_in_dienst.append('No')

# Add the lists as new columns in sales_staff
sales_staff['KORT_IN_DIENST'] = kort_in_dienst
sales_staff['LANG_IN_DIENST'] = lang_in_dienst

# Display the DataFrame with only 'LAST_NAME', 'KORT_IN_DIENST', and 'LANG_IN_DIENST' columns
overview = sales_staff[['LAST_NAME', 'KORT_IN_DIENST', 'LANG_IN_DIENST']]

overview
```
